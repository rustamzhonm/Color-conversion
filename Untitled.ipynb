```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
import time
import scipy.io as sp

```

```python
asim_data = sp.loadmat("asimdata.mat")
print(asim_data.keys())

abs_green_GB = asim_data.get("GBgreenabs")
abs_green_RGB = asim_data.get("RGBgreenabs")
abs_red_RGB = asim_data.get("RGBorangeabs")

PL_green_GB = asim_data.get("GBgreenpl")
PL_green_RGB = asim_data.get("RGBgreenpl")
PL_red_RGB = asim_data.get("RGBorangepl")

blue5 = asim_data.get("blue5")

Q = asim_data.get("Q")
Q = np.concatenate(Q,axis = 0)

w = np.arange(400,801)

absorbance = np.array([abs_green_GB[:,1]-min(abs_green_GB[:,1]), abs_green_RGB[:,1]-min(abs_green_RGB[:,1]),abs_red_RGB[:,1]-min(abs_red_RGB[:,1])])
PL = np.array([PL_green_GB[:,1]-min(PL_green_GB[:,1]), PL_green_RGB[:,1]-min(PL_green_RGB[:,1]),PL_red_RGB[:,1]-min(PL_red_RGB[:,1])])

PL[0,:] = PL[0,:]/sum(PL[0,:])
PL[1,:] = PL[1,:]/sum(PL[1,:])
PL[2,:] = PL[2,:]/sum(PL[2,:])
PL = PL.T
absorbance = absorbance.T

OD = np.array([abs_green_GB[60,1], abs_green_RGB[60,1], abs_red_RGB[60,1]])

blue5 = blue5[:,1]/sum(blue5[:,1])

alphaa = 2.3*absorbance[:,[0, 2]]
Pa = PL[:,[0, 2]]
QQ = Q[[0, 2]]
QQ = np.array([1, 1])
ODD = OD[[0, 2]]

l = len(QQ.T)

XX = 1./ODD

x = np.linspace(0,XX[0],10)
y = np.linspace(0,XX[1],10)

count = 0
alphaa1 = np.zeros(alphaa.shape)
Paa = np.zeros(Pa.shape)
W = np.zeros(QQ.shape)
S_out = np.zeros((len(w),len(x),len(y)))
EQE = np.zeros((len(x),len(y)))


for i in range(len(x)):
    for j in range(len(y)):
        
        D = np.array([x[i], y[j]])
        
        alphaa1[:,0] = alphaa[:,0]*D[0]
        alphaa1[:,1] = alphaa[:,1]*D[1]

        
        K[:,0] = alphaa1[:,0]/(alphaa1[:,0]+alphaa1[:,1]+0.000001)
        K[:,1] = alphaa1[:,1]/(alphaa1[:,0]+alphaa1[:,1]+0.000001)

        K[np.isnan(K)] = 0
        
        for k in range(l):
            Paa[:,k] = Pa[:,k]/sum(Pa[:,k])
            
        A = np.exp(-(alphaa1[:,0] + alphaa1[:,1]))
        E = 1 - A
        C = np.zeros((l,l))
        
        for ii in range(l):
            for jj in range(l):
                C[ii,jj] = QQ[jj]*sum(Paa[:,ii]*K[:,jj]*E)
        
        Ci = np.linalg.inv(np.eye(l)-C)
        
        for ii in range(l):
            W[ii] = QQ[ii]*sum(blue5*K[:,ii]*E)
        
        AE = np.matmul(np.matmul(Paa,Ci.T),W)
        S_out[:,i,j] = (blue5 + AE)*A
        EQE[i,j] = 100*sum(AE*A) / sum((blue5))

        plt.plot(w,S_out[:,i,j])
        #plt.plot(w,E)
plt.show()
EQE
```
